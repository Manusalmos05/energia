```python
# 🤖 Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import streamlit as st
```

```python
# Cargar el archivo CSV en un DataFrame y dejar columnas 'fecha' y 'demanda'
ruta_csv = '../data/raw/entrenamiento/Evolución de la demanda (B.C.)_01-01-2021_31-12-2024.csv'
df = pd.read_csv(ruta_csv, delimiter=',')
df = df.T.reset_index()
df.columns = ['fecha', 'demanda']
df = df.drop(df.index[0])  # Elimina la fila de encabezado transpuesta
df.head()
```

```python
df.info()
```

```python
# Limpiar y convertir la columna 'fecha' a tipo datetime, luego a año/mes
fechas = pd.to_datetime(df['fecha'], errors='coerce', utc=True)
df = df[fechas.notna()].copy()
df['fecha'] = fechas[fechas.notna()].dt.strftime('%Y/%m')
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y/%m')
df.head()
```

```python

# Convertir la columna 'demanda' a float con 6 decimales
df['demanda'] = pd.to_numeric(df['demanda'], errors='coerce').round(6)
```
